```python
import numpy as np
import matplotlib.pyplot as plt

def draw_sigmoid(start, end):
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10, 2))

    x = np.linspace(start, end, 1000)
    sigmoid = lambda v: 1. / (1. + np.exp(-v))

    sigmoid_vals = sigmoid(x)
    sigmoid_deriv = sigmoid_vals * (1 - sigmoid_vals)

    ax1.plot(x, sigmoid_vals)
    ax1.set_title("$\sigma(x)$")
    ax1.set_xlabel("$x$")
    ax2.plot(sigmoid_vals, sigmoid_deriv)
    ax2.set_title("$\sigma(x)^{'}$ on $\sigma(x)$")
    ax2.set_xlabel("$\sigma(x)$")
    ax3.plot(x, sigmoid_deriv)
    ax3.set_title("$\sigma(x)^{'}$ on $x$")
    ax3.set_xlabel("$x$")

    plt.show()

draw_sigmoid(-50, 50)
draw_sigmoid(-8, 8)
```
